```python
files=[]
for c in categoryNames:
    img_file=glob.glob("Dataset/"+c+"/"+"*.jpg")
    files.append(img_file)
    print(len(img_file))
```

```python
for i in range(len(categoryNames)):
        if not os.path.exists("C:/Users/ebrut/Desktop/Son/Dataset-withoutbackground/"+categoryNames[i]):
         os.mkdir("Dataset-withoutbackground/"+categoryNames[i])
```

```python
for c in categoryNames:
    index=categoryNames.index(c)
    for file in files[index]:
        input_p=file
        print(file)
        output_p=input_p.replace("C:/Users/ebrut/Desktop/Son/Dataset/"+c,"C:/Users/ebrut/Desktop/Son/Dataset-withoutbackground/"+c)
        with open(input_p,"rb") as i:
            with open(output_p,"wb") as o:
                ori=i.read()
                out=remove(ori)
                o.write(out)
```
